```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
import pandas as pd
df = pd.read_csv("CarPrice_Assignment.csv",index_col=0)
df = df.dropna()
Car = pd.DataFrame(df)
```

```python
Car.head()
```

```python
Car.corr()["price"].sort_values()
```

```python
X = Car["horsepower"].values.reshape(-1,1)
Y = Car["price"].values.reshape(-1,1)

plt.scatter(X,Y)
plt.xlabel("horsepower")
plt.ylabel("price");
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3)




reg = LinearRegression().fit(X_train, Y_train )


y_train_hat = reg.predict(X_train)
y_test_hat = reg.predict(X_test)
```

```python
plt.scatter(X_train, Y_train,label = "train")
plt.scatter(X_test, Y_test,label = "test")
plt.xlabel("horsepower")
plt.ylabel("price");
plt.legend();
```

```python
plt.scatter(X_train, Y_train)
plt.scatter(X_test, Y_test)
X_plot = np.linspace(50,300).reshape(-1,1)
Y_plot = reg.predict(X_plot)
#GrÃ¡fica del modelo
plt.plot(X_plot,Y_plot, color = "red", label = "Modelo")
plt.xlabel("horsepower")
plt.ylabel("price");
plt.legend();
```

```python
print("Entrenamiento : ",r2_score(Y_train,y_train_hat))
print("Prueba : ",r2_score(Y_test,y_test_hat))
```
